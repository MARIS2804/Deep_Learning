```python
X = np.array([[0, 0],
              [0, 1],
              [1, 0],
              [1, 1]])

O = np.array([[0],
              [1],
              [1],
              [0]])

input_hidden_weights = [[-23.0,11],[-33,-22]]
hidden_output_weights = [[-22.0],[14.0]]
```

```python
import math
def sigmoid(Xw):
    for i in range(0,len(Xw)):
        val=1/(1+(math.exp(-Xw[i])))
        if(val>=0.5):
            Xw[i]=1
        else:
            Xw[i]=0
    return Xw
```

```python
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
lr = 0.5
for iteration in range(2000):
    print("Itearation =",iteration+1)
    for i in range(4):
        input_to_hidden = np.dot(X[i], input_hidden_weights)
        input_to_hidden = input_to_hidden.reshape(-1, 1)
        #print(input_to_hidden)
        hidden_output = sigmoid(input_to_hidden)
        input_to_output_neuron = np.dot(hidden_output.T, hidden_output_weights)
        ypred = sigmoid(input_to_output_neuron)
        error_at_output_layer = O[i] - ypred
        print(error_at_output_layer)
        if np.all(error_at_output_layer) == 0:
            continue
        delta_at_output_layer_neuron = error_at_output_layer*sigmoid_derivative(ypred)
        hidden_output_weights += lr*np.dot(hidden_output, delta_at_output_layer_neuron)
        delta_at_hidden_layer_neurons = np.dot(hidden_output_weights, delta_at_output_layer_neuron)*sigmoid_derivative(hidden_output)
        input_hidden_weights += lr*np.dot(X[i].reshape(-1, 1), delta_at_hidden_layer_neurons.T).T
        #print(ypred)
    print()
```
