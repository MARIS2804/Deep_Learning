```python
import math
def step_function(Xw):
    for i in range(0,len(Xw)):
        if(Xw[i]<=5):
            Xw[i]=0
        else:
            Xw[i]=1
    return Xw

def relu(Xw):
    for i in range(0,len(Xw)):
        if(Xw[i]<=0):
            Xw[i]=0
    return Xw

def sigmoid(Xw):
    for i in range(0,len(Xw)):
        val=1/(1+(math.exp(-Xw[i])))
        if(val>0.5):
            Xw[i]=1
        else:
            Xw[i]=0
    return Xw
```

```python
#AND
import numpy as np
import matplotlib.pyplot as plt
X=np.array([[0,0],[0,1],[1,0],[1,1]])
o=np.array([0,0,0,1])
w=np.array([-23.0,15.0])
bias=-10
lr=0.8
itr=1
xi=[i[0] for i in X]
yi=[i[1] for i in X]
plt.scatter(xi,yi)
w1=[]
w2=[]
while(1):
    #print("EPOCHS ",itr)
    itr+=1
    Xw=np.dot(X,w)
    Xw=np.add(Xw,bias)
    Xw=sigmoid(Xw)
    
    loss_vec=np.subtract(o,Xw)
    # print("Xw= ",Xw)
    # print("w= ",w)
    # print("bias= ",bias)
    # print("Loss vector = ",loss_vec)
    # print()
    w1.append(w[0])
    w2.append(w[1])
    if(np.any(loss_vec)):
        for i in range(0,4):
            if(loss_vec[i]!=0):
                change=lr*loss_vec[i]*X[i][0]
                w[0]+=change
                change=lr*loss_vec[i]*X[i][1]
                w[1]+=change
                bias=bias+(lr*loss_vec[i])
    else:
        break
    
x = np.linspace(0, 1, 100)
y = -(w[0] / w[1]) * x - bias / w[1]
plt.plot(x, y,'black')
    
      
```
